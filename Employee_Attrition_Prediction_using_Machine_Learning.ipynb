# Task 4: Feature Engineering & Data Preprocessing

In this section, we will prepare the dataset for machine learning by:

- Separating features and target variable
- Applying One-Hot Encoding to categorical features
- Splitting the dataset into training and testing sets
- Scaling numerical features using StandardScaler

```python
# Features and Target

X = df.drop("Attrition", axis=1)
y = df["Attrition"]

print("Features Shape :", X.shape)
print("Target Shape :", y.shape)
```

```python
num_cols = X.select_dtypes(include=["int64", "float64"]).columns

cat_cols = X.select_dtypes(include="object").columns

print("Numerical Features :", len(num_cols))
print("Categorical Features :", len(cat_cols))
```

```python
X = pd.get_dummies(
    X,
    columns=cat_cols,
    drop_first=True
)

print("Shape after Encoding :", X.shape)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("Training Data :", X_train.shape)
print("Testing Data :", X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train[num_cols] = scaler.fit_transform(X_train[num_cols])

X_test[num_cols] = scaler.transform(X_test[num_cols])
```

```python
X_train.head()
```

```python
print("Training Set")

print(y_train.value_counts(normalize=True) * 100)

print("\nTesting Set")

print(y_test.value_counts(normalize=True) * 100)
```

```python
feature_names = X.columns.tolist()

print("Total Features after Encoding :", len(feature_names))
```

```python
feature_names[:20]
```

## Observations

- The target variable was separated from the input features.
- All categorical columns were converted into numerical format using One-Hot Encoding.
- The dataset was split into training and testing sets in an 80:20 ratio.
- Numerical features were standardized using StandardScaler to improve model performance.
- The class distribution remained consistent across both training and testing datasets due to stratified sampling.

# Task 4: Model Building & Comparison

In this section, we will train three machine learning models:

- Logistic Regression
- Random Forest Classifier
- Gradient Boosting Classifier

The models will be compared using different evaluation metrics to identify the best performing model.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score
)
```

```python
models = {
    "Logistic Regression": LogisticRegression(
        class_weight="balanced",
        random_state=42,
        max_iter=1000
    ),

    "Random Forest": RandomForestClassifier(
        n_estimators=200,
        class_weight="balanced",
        random_state=42
    ),

    "Gradient Boosting": GradientBoostingClassifier(
        random_state=42
    )
}
```

```python
results = []

for name, model in models.items():

    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    y_prob = model.predict_proba(X_test)[:,1]

    accuracy = accuracy_score(y_test, y_pred)

    precision = precision_score(y_test, y_pred)

    recall = recall_score(y_test, y_pred)

    f1 = f1_score(y_test, y_pred)

    roc = roc_auc_score(y_test, y_prob)

    results.append([
        name,
        accuracy,
        precision,
        recall,
        f1,
        roc
    ])
```

```python
comparison = pd.DataFrame(
    results,
    columns=[
        "Model",
        "Accuracy",
        "Precision",
        "Recall",
        "F1 Score",
        "ROC-AUC"
    ]
)

comparison
```

```python
comparison.iloc[:,1:] = comparison.iloc[:,1:].round(3)

comparison
```

```python
best_model_name = comparison.sort_values(
    by="ROC-AUC",
    ascending=False
).iloc[0]["Model"]

print("Best Model :", best_model_name)
```

```python
best_model = models[best_model_name]
```

```python
comparison.sort_values(
    by="ROC-AUC",
    ascending=False
).reset_index(drop=True)
```

## Model Comparison

The three machine learning models were trained and evaluated on the test dataset.

The comparison table shows the Accuracy, Precision, Recall, F1 Score, and ROC-AUC Score for each model.

The model with the highest ROC-AUC score will be selected for further evaluation.

# Bonus: Hyperparameter Tuning

Instead of using the default model parameters, we will try different parameter combinations using GridSearchCV to improve model performance.

Hyperparameter tuning helps in finding the best model configuration.

```python
from sklearn.model_selection import GridSearchCV
```

```python
lr_params = {
    "C": [0.01, 0.1, 1, 10],
    "solver": ["liblinear", "lbfgs"]
}

lr_grid = GridSearchCV(
    LogisticRegression(
        class_weight="balanced",
        random_state=42,
        max_iter=1000
    ),
    lr_params,
    cv=5,
    scoring="f1",
    n_jobs=-1
)

lr_grid.fit(X_train, y_train)

print("Best Parameters:")
print(lr_grid.best_params_)

print("\nBest F1 Score:")
print(round(lr_grid.best_score_,3))
```

```python
rf_params = {
    "n_estimators": [100,200,300],
    "max_depth": [5,10,None],
    "min_samples_split": [2,5],
    "min_samples_leaf": [1,2]
}

rf_grid = GridSearchCV(
    RandomForestClassifier(
        class_weight="balanced",
        random_state=42
    ),
    rf_params,
    cv=5,
    scoring="f1",
    n_jobs=-1
)

rf_grid.fit(X_train, y_train)

print("Best Parameters:")
print(rf_grid.best_params_)

print("\nBest F1 Score:")
print(round(rf_grid.best_score_,3))
```

```python
gb_params = {
    "n_estimators": [100,200],
    "learning_rate": [0.01,0.05,0.1],
    "max_depth": [3,5]
}

gb_grid = GridSearchCV(
    GradientBoostingClassifier(
        random_state=42
    ),
    gb_params,
    cv=5,
    scoring="f1",
    n_jobs=-1
)

gb_grid.fit(X_train, y_train)

print("Best Parameters:")
print(gb_grid.best_params_)

print("\nBest F1 Score:")
print(round(gb_grid.best_score_,3))
```

```python
tuned_results = pd.DataFrame({
    "Model":[
        "Logistic Regression",
        "Random Forest",
        "Gradient Boosting"
    ],
    "Best CV F1 Score":[
        lr_grid.best_score_,
        rf_grid.best_score_,
        gb_grid.best_score_
    ]
})

tuned_results["Best CV F1 Score"] = tuned_results["Best CV F1 Score"].round(3)

tuned_results.sort_values(
    by="Best CV F1 Score",
    ascending=False
)
```

```python
scores = {
    "Logistic Regression": lr_grid.best_score_,
    "Random Forest": rf_grid.best_score_,
    "Gradient Boosting": gb_grid.best_score_
}

best_tuned_model = max(scores, key=scores.get)

print("Best Tuned Model :", best_tuned_model)
```

```python
if best_tuned_model == "Logistic Regression":
    final_model = lr_grid.best_estimator_

elif best_tuned_model == "Random Forest":
    final_model = rf_grid.best_estimator_

else:
    final_model = gb_grid.best_estimator_
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

y_pred = final_model.predict(X_test)
y_prob = final_model.predict_proba(X_test)[:,1]

print("Accuracy :", round(accuracy_score(y_test,y_pred),3))
print("Precision :", round(precision_score(y_test,y_pred),3))
print("Recall :", round(recall_score(y_test,y_pred),3))
print("F1 Score :", round(f1_score(y_test,y_pred),3))
print("ROC AUC :", round(roc_auc_score(y_test,y_prob),3))
```

```python
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6,5))

sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues"
)

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix (Best Tuned Model)")

plt.show()
```

```python
if hasattr(final_model, "feature_importances_"):

    importance = pd.DataFrame({
        "Feature": X.columns,
        "Importance": final_model.feature_importances_
    })

else:

    importance = pd.DataFrame({
        "Feature": X.columns,
        "Importance": abs(final_model.coef_[0])
    })

importance = importance.sort_values(
    by="Importance",
    ascending=False
)

importance.head(10)
```

```python
plt.figure(figsize=(8,6))

sns.barplot(
    data=importance.head(10),
    x="Importance",
    y="Feature"
)

plt.title("Top 10 Important Features")

plt.show()
```

# Explainable AI

Machine Learning models often make predictions that are difficult to interpret.

To understand which features influence employee attrition the most, we will use feature importance and permutation importance.

```python
from sklearn.inspection import permutation_importance
```

```python
result = permutation_importance(
    final_model,
    X_test,
    y_test,
    n_repeats=10,
    random_state=42
)

perm_importance = pd.DataFrame({
    "Feature": X.columns,
    "Importance": result.importances_mean
})

perm_importance = perm_importance.sort_values(
    by="Importance",
    ascending=False
)

perm_importance.head(10)
```

```python
plt.figure(figsize=(10,6))

sns.barplot(
    data=perm_importance.head(10),
    x="Importance",
    y="Feature"
)

plt.title("Top 10 Important Features (Permutation Importance)")

plt.show()
```

## Observation

Permutation Importance shows which employee attributes contribute the most towards predicting attrition.

Features with higher importance have a greater impact on the model's prediction.
